```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
# COUNTPLOTS :  First step to do with the countplots
import seaborn as sns
sns.countplot(x = 'Survived' , data = df)
df['Survived'].value_counts()
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].value_counts().plot(kind='bar')
```

```python
df['Embarked'].value_counts().plot(kind='pie', autopct='%1.0f%%')  #We can search documentations to show percentages on pie chart or etc from google.
```
